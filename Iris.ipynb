```python
data1 = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
target1 = pd.DataFrame(iris.target, columns=['species'])
df = pd.concat([data1, target1], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=1)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
data = pd.DataFrame(X_train, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
data.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier().fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
score = knn.score(X_test, y_test)
print('Score : {:.2f}'.format(score))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(150, random_state=1).fit(X_train, y_train)
```

```python
pred = rfc.predict(X_test)
score = rfc.score(X_test, y_test)
print('Score : {:.2f}'.format(score))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C=1.5, gamma='auto', random_state=1).fit(X_train, y_train)
pred = svc.predict(X_test)
score = svc.score(X_test, y_test)
print('Score : {:.2f}'.format(score))
```
